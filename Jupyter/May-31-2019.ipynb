## Ilustrate 1D Regression Tree

```python
rawdata=pd.read_csv("short_flux_data_file.csv")
rawdata[:5]
```

```python
X=np.array(rawdata["date"]).reshape(-1,1)
Y=np.array(rawdata["flux"]).reshape(-1,1)
plt.scatter(X,Y, marker="+",color="magenta")
plt.xlabel("Days into data file")
plt.ylabel("10.7 cm Solar Flux");
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
r2 = DecisionTreeRegressor(max_depth=2)
r2.fit(X,Y)
xplot=np.linspace(0,30,200).reshape(-1,1)
Y2=r2.predict(xplot)

r4 = DecisionTreeRegressor(max_depth=4)
r4.fit(X,Y)
Y4=r4.predict(xplot)

r7= DecisionTreeRegressor(max_depth=7)
r7.fit(X,Y)
Y7=r7.predict(xplot)
```

```python
X=np.array(rawdata["date"]).reshape(-1,1)
Y=np.array(rawdata["flux"]).reshape(-1,1)
plt.scatter(X,Y, marker="d",color="blue",label="data")

plt.plot(xplot,Y2,c="red",label="depth 2 tree")
plt.plot(xplot,Y4,c="orange",label="depth 4 tree")
plt.plot(xplot,Y7,c="black",label="depth 7 tree")


plt.legend()

plt.xlabel("Days into data file")
plt.ylabel("10.7 cm Solar Flux");
```

```python
from sklearn.metrics import mean_squared_error
for depth in range(2,9):
    model = DecisionTreeRegressor(max_depth=depth)
    model.fit(X,Y)
    YP=model.predict(X)
    MSE=mean_squared_error(YP,Y)
    print("The MSE for a depth of",depth,"is",MSE)
```

Old function from sklearn help files

```python
def find_plateau_centers(Z,xmin,xmax,ymin,ymax):
    U=np.unique(Z)
    centers=[]
    for x in U:
        locs=[]
        for ix in np.arange(xmin,xmax,1):
            for iy in np.arange(ymin,ymax,1):
                zval=Z[int(iy-ymin)][int(ix-xmin)]
                if zval==x:
                    locs.append((ix,iy))
        center=list(np.mean(locs,axis=0))
        centers.append((x,center))
    return(centers)
```

```python
plt.scatter(displacement,horsepower)
xmin=displacement.min()-1
xmax=displacement.max()+1
ymin=horsepower.min()-1
ymax=horsepower.max()+1
plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax)

xx,yy=np.meshgrid(np.arange(xmin,xmax,1), np.arange(ymin,ymax,1))
ZP=model.predict(np.c_[xx.ravel(),yy.ravel()])
Z=ZP.reshape(xx.shape)
plt.contour(xx,yy,Z, colors="black")

centers = find_plateau_centers(Z,xmin,xmax,ymin,ymax)
for value, location in centers:
    plt.text(*location, str(round(value,2)), fontsize=15, 
             horizontalalignment="center")


plt.xlabel("Displacement")
plt.ylabel("Horsepower")
plt.gcf().set_size_inches(6,6)
```

Classification Tree

```python
X=np.array(data[["mpg","displ","hp","weight","accel"]])
```

```python
cylinders=np.array(data["cyl"])
classvals=[0,0,1,1,2]
unis=np.unique(cylinders).tolist()
print(unis)
```

```python
Y=np.array([classvals[unis.index(x)] for x in cylinders])
print(Y)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
nsplits=1000
depth=3
errs=[]
for j in range(nsplits):
    xtrain,xtest,ytrain,ytest=train_test_split(X,Y)
    DT=DecisionTreeClassifier(max_depth=depth)
    DT.fit(xtrain,ytrain)
    YP=DT.predict(xtest)
    errs.append(1-accuracy_score(ytest,YP))
```

```python
np.mean(errs), np.std(errs)
```

```python
plt.hist(errs, edgecolor="k",color="red");
```
