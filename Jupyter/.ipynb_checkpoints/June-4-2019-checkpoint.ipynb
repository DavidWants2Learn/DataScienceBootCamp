```python
from sklearn.decomposition import PCA 
# PCA means Princ. Component Analysis
```

```python
X=np.array(cloud)
X[:3] # show first 3 rows
```

```python
pca=PCA(n_components=2)
pca.fit(X)
print("components=")
print(pca.components_)
```

```python
print("Variance ratio:", pca.explained_variance_ratio_)
```

```python
P=pca.transform(X)
P[:3]
```

```python
xP=P[:,0] # first column of P
yP=P[:,1] # second column of P
```

```python
plt.scatter(xP,yP,marker=".",c="red")
plt.scatter(x,y,marker=".",c="k")
plt.title("Both Clouds")
plt.xlim(-7,15); plt.ylim(-7,15); 
plt.grid(); plt.axhline(y=0,c="k"); plt.axvline(x=0,c="k")
plt.gcf().set_size_inches(5,5)
```

```python
data=pd.read_csv("cardata.csv")
data[:5]
```

```python
# cut and paste from May 31
cylinders=np.array(data["cyl"])
classvals=[0,0,1,1,2]
unis=np.unique(cylinders).tolist()
print(unis)
Y=np.array([classvals[unis.index(x)] for x in cylinders])
print(Y)
```

```python
DF=pd.get_dummies(data,columns=["origin"])
DF=DF.drop(columns=["cyl"])
DF[:5]
```

```python
X=np.array(DF)
X[:5]
```

```python
pca=PCA(n_components=8)
pca.fit(X)
pca.explained_variance_ratio_
```

```python
for value in pca.explained_variance_ratio_:
    print("%.15f percent" %(100*value))
```

```python
pca=PCA(n_components=2)
pca.fit(X)
pca.explained_variance_ratio_
```

```python
P=pca.transform(X)
```

```python
colors=["cyan","magenta","purple"]
markers=["+","D","."]
for i in range(3):
    x,y=zip(*P[Y==i])
    plt.scatter(x,y,marker=markers[i],c=colors[i],s=10)
plt.grid()
plt.gcf().set_size_inches(5,5)
plt.gca().set_facecolor("lightyellow")
```

```python
from sklearn.neural_network import MLPClassifier as ANN
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
r=ANN(solver="lbfgs", alpha=1e-5, random_state=1,
     hidden_layer_sizes=(10,10,10))
Z=scaler.fit_transform(P)
nruns=100
accs=[]
for j in range(nruns):
    ztrain,ztest,ytrain,ytest=train_test_split(Z, Y)
    r.fit(ztrain, ytrain)
    YP=r.predict(ztest)
    accs.append(accuracy_score(YP,ytest))
print(np.mean(accs), np.std(accs))
```

```python
plt.hist(accs, color="lightblue", edgecolor="k", bins=5);
```

```python
def visualize_clasifier_contours(clf,X,Y,DX,DY,colors,markers,xlab="X",
                                thelevels=[.5,1.5,2.5,3.5],
                                ylab="Y",xlim=[-.1, 1.1],ylim=[-.1, 1.1]):
    xinterval = X[:, 0].min() - 1, X[:, 0].max() + 1
    yinterval = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(*xinterval, DX),
                         np.arange(*yinterval, DY))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    c=plt.contour(xx, yy, Z, alpha=1,levels=thelevels,linewidths=2,colors="k")
    c=plt.contourf(xx, yy, Z, alpha=.1)

  
    nclasses=len(colors)
    for i in range(nclasses):
        #x,y=zip(*X[Y==i]) 
        x,y=(X[Y==i].T).tolist() # is equivalent to x,y=zip(*X[Y==i])
        plt.scatter(x,y,marker=markers[i],color=colors[i],s=5)

    plt.xlabel(xlab,fontsize=14)
    plt.ylabel(ylab,fontsize=14)
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.grid()
    fig=plt.gcf()
    fig.tight_layout()
    fig.set_size_inches(6,6)
```

```python
visualize_clasifier_contours(r,Z,Y,.01,.01, colors, markers)
```

# Neat Image Compression Example

```python
from PIL import Image
```

```python
g=Image.open("BUILDING.JPG")
plt.imshow(g);
```

```python
G=g.convert("L") # create a gray scale image
A=np.asarray(G)  # converting to a numpy array
plt.matshow(A, cmap="gray");
```

```python
A
```

```python
A.shape
```

```python
U, Sigma, V=np.linalg.svd(A) # Singular Value Decomposition
```

```python
Sigma.shape
```

Compute Principal Component Expansion as per Theorem B.13 of Getting Started in Machine Learning

```python
UT=U.T
def PC(n): # this returns an image with the first n princ. components
    M = np.zeros_like(A).astype(float)
    for j in range(n):
        B=np.outer(UT[j], V[j])*Sigma[j]
        M+=B
    return(M)
```

```python
plt.imshow(PC(400),cmap="gray")
```
