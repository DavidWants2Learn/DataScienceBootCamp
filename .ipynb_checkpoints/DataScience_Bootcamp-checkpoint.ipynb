```python
#Question 1 word count

f=open("Excercises/bobby.txt", "r")

count = {}

if f.mode == 'r':
    contents =f.read()
    print(contents)
    for w in open("Excercises/bobby.txt").read().split():
        if w in count:
            count[w] += 1
        else:
            count[w] = 1
    for word, times in count.items():
        print ("%s = %d" % (word, times))
```

```python
#Question 2 GCD

def GCD(a, b):
    while (b >= 0):
        if (b == 0):
            return a
        else:
            if (a > b):
                a -= b
            elif (a < b):
                r = a
                a = b
                b = r
            else:
                a = b
                b = 0
        print(a,b)

GCD(330, 156)
#GCD(25963443,17157)
```

```python
#Question 3 Sieve of Eratosthones

prime = []

def func(n):
    for num in range(n):
        if ( num==1 or num==2 or num==3 ):
            prime.append(num)
        
func(10)
print(prime)
```
